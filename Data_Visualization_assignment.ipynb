```python
# 3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.


import pandas as pd

# Create the dataset
data = {
    'Category': ['Electronics', 'Clothing', 'Groceries', 'Furniture', 'Toys', 'Books', 'Health & Beauty'],
    'Value': [1500, 800, 300, 1200, 400, 600, 500]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate total value
total_value = df['Value'].sum()

# Calculate average value
average_value = df['Value'].mean()

# Find maximum and minimum values
max_value = df['Value'].max()
min_value = df['Value'].min()

# Find the categories with max and min values
max_category = df.loc[df['Value'] == max_value, 'Category'].values[0]
min_category = df.loc[df['Value'] == min_value, 'Category'].values[0]

# Calculate percentage contribution of each category
df['Percentage'] = (df['Value'] / total_value) * 100

# Print results
print("Total Value:", total_value)
print("Average Value:", average_value)
print("Maximum Value:", max_value, "in", max_category)
print("Minimum Value:", min_value, "in", min_category)
print("\nPercentage Contribution:\n", df[['Category', 'Value', 'Percentage']])
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n_samples = 100
feature_1 = np.random.normal(0, 1, n_samples)
feature_2 = feature_1 + np.random.normal(0, 0.1, n_samples)  # Correlated with feature_1
feature_3 = feature_1 * 0.5 + np.random.normal(0, 0.5, n_samples)  # Correlated with feature_1
feature_4 = np.random.normal(0, 1, n_samples)  # Uncorrelated feature

# Create a DataFrame
data = {
    'Feature_1': feature_1,
    'Feature_2': feature_2,
    'Feature_3': feature_3,
    'Feature_4': feature_4
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})

# Title and labels
plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)


# Import necessary libraries
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.layouts import column

# Set up output in a Jupyter notebook
output_notebook()

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)

# Create x and y coordinates
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare the data for Bokeh
# Flatten the arrays for Bokeh
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, z=z_flat))

# Create a Bokeh figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=600, height=600)

# Create a color map
mapper = linear_cmap(field_name='z', palette='Viridis256', low=min(z_flat), high=max(z_flat))

# Add squares (rectangles) to the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       fill_color=mapper, line_color='black')

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)

```
